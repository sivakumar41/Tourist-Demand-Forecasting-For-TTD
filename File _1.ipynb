```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import numpy as np
kf = KFold(n_splits=6, shuffle=True, random_state=815)  
model2=lgb.LGBMRegressor()
accuracies = []  
for train_idx, test_idx in kf.split(x):
    x_train1, y_train1 = x[train_idx], y[train_idx]  # Get the training data for this fold
    x_test1, y_test1 = x[test_idx], y[test_idx]  # Get the test data for this fold

    # Train your model on the training data
    model2.fit(x_train1, y_train1)

    # Make predictions on the test data and calculate accuracy
    y_pred1 = model2.predict(x_test1)
    accuracy21=model2.score(x_test1,y_test1)
    print("The  testing accuracy is ",accuracy21)
    accuracies.append(accuracy21)

mean_accuracy = np.mean(accuracies) # Calculate the mean accuracy across all folds
print("mean_accuracy is",mean_accuracy)
```
