```python
sno=df['Unnamed: 0'].tolist()
l=['Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday']
```

```python
update={'Sunday':[[],[]],'Monday':[[],[]],'Tuesday':[[],[]],'Wednesday':[[],[]],'Thursday':[[],[]],'Friday':[[],[]],'Saturday':[[],[]]}
```

```python
for i in (sno):
    update[(l[i%7])][0]+=[pil[i]]
    update[(l[i%7])][1]+=[ton[i]]
```

```python
average={'Sunday':[],'Monday':[],'Tuesday':[],'Wednesday':[],'Thursday':[],'Friday':[],'Saturday':[]}
```

```python
for i in update:
    ap=update[i][0]
    at=update[i][1]
    sp=0
    np=0
    st=0
    nt=0
    for j in ap:
        if str(j).isdigit():
            sp+=j
            np+=1
    for k in at:
        if str(k).isdigit():
            st+=k
            nt+=1
    average[i]=[sp//np,st//nt]
average
```
